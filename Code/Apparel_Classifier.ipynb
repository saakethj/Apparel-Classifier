# Apparel Classifier
This *dataset* contains **10,714** grayscale images in 4 categories with (28x28) resolution. Here, **7,400** images are used to train the model and **3,314** images are used to evaluate how accurately the network learned.

The 4 categories in the dataset are:


*   **Footwear**
*   **Handbags**
*   **Shirts**
*   **T-shirts**

Each category corresponds to respective Lables. These Labels are of integer datatype and stored in array.


```python
category_names = ['Footwear', 'Handbags', 'Shirts', 'T-shirts']
```

```python
X_train_images = np.load('Train images.npz')
Y_train_labels = np.load('Train labels.npz')
X_test_images = np.load('Test images.npz')
Y_test_labels = np.load('Test labels.npz')
```

```python
train_images = X_train_images['X_train']
train_labels = Y_train_labels['Y_train']
test_images = X_test_images['X_test']
test_labels = Y_test_labels['Y_test']
```

**Dimensions and length of Train and Test dataset**


The Train dataset has total of 7,400 images and stored in 2-D array with pixel values ranging from 0-255.The train labels are of length same as train images.

The Test datset has total of 3314 images with corresponding labels.
